```python
import pandas as pd
import numpy as np
df = pd.read_csv("data.csv")
```

```python
for row in range(len(df)):
    #print(len(df.loc[row, "Game Data"]))
    df.at[row, 'Game Data'] = df.loc[row, "Game Data"].split("Score By")
    #print(len(df.loc[row, "Game Data"]))
```

```python
for row in range(len(df)):
    newGameData = []
    for r in df.loc[row, "Game Data"]:
        newGameData.append(r.replace('\\n', '').replace('\\t', ''))
    df.at[row, 'Game Data'] = newGameData
```
